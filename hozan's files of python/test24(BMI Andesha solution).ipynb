```python
x=0
def f():
    global x
    x=1
    
    a.set('Standard')
    aa.set('weight in pound')
    aaa.set('height in feet')
def d():
    if x==0:
        m=(float(e.get())/(float(ee.get())**2))
        c.set(f'{m}')
        if m>18.5:
            cc.set("underrrrrr in metric")
        if m>=18.5 and m<= 25.0:
            cc.set("normallll in metric")
        if m>=25.0 and m<=29.9:
            cc.set("oveeeeeeer in metric")
        if m>30:
            cc.set("pakuuuuuuuuf in metric")
    else:
        s=((float(e.get())/(float(ee.get())**2))*703)
        c.set(f'{s}')
        if s>18.5:
            cc.set("underrrrrr in standars")
        if s>=18.5 and s<= 25.0:
            cc.set("normallll in standars")
        if s>=25.0 and s<=29.9:
            cc.set("oveeeeeeer in standard")
        if s>30:
            cc.set("pakuuuuuuuuf standard")

from tkinter import *
r=Tk()
r.geometry('300x300')
r.configure(bg='black')
r.title("bmi")

a=StringVar()
a.set('METRIC')
b=Button(r,textvariable=a,bg='red',command=f)
b.pack()

aa=StringVar()
aa.set('weight in KG')
ll=Label(r,textvariable=aa,bg='black',fg='white')
ll.pack()
       

e=Entry(r)
e.pack()

aaa=StringVar()
aaa.set("height in meter")
lll=Label(r,textvariable=aaa,bg='black',fg='white')
lll.pack()

ee=Entry(r)
ee.pack()

bb=Button(r,text="CALCULATE",command=d)
bb.pack()

c=StringVar()
c.set("")
lo=Label(r,textvariable=c)
lo.pack()


cc=StringVar()
cc.set("")
laa=Label(r,textvariable=cc)

laa.pack()

r.mainloop()
```
