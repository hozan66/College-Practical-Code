```python
def d1():
    v1.set("Metric")
    l1.config(command=d)
    v3.set("in M")
    v4.set("Weight in Kg")
def d():
    v1.set("standard")
    l1.config(command=d1)
    v3.set("in inch")
    v4.set("Weight in Ibs")
def d2():
    if(v1.get()=="standard"):
        v6.set((v5.get()/(v2.get()**2))*703)
        
    elif(v1.get()=="Metric"):
        v6.set((v5.get()/(v2.get()**2)))
    if(v6.get()<18.5):
        v7.set("under")
        image.config(image=p1)
    elif(v6.get()>=18.5 and v6.get()<25):
        v7.set("normal")
        image.config(image=p2)
    elif(v6.get()>=25 and v6.get()<29.9):
        v7.set("over")  
        image.config(image=p3)
    else:
        
        v7.set("Super Obese")  
        image.config(image=p4)
    re1.pack()   
    image.pack()
    
    
from tkinter import *
r1=Tk()
r1.config(bg="black")
v1=StringVar()
v1.set("Metric")
l1=Checkbutton(r1,bg="yellow",textvariable=v1,indicatoron=0,command=d)
l1.pack()
 
v3=StringVar()
v3.set("in M")
l2=Label(r1,textvariable=v3)
l2.pack()
 
v2=DoubleVar()
e1=Entry(r1,textvariable=v2,bg="black",foreground="red")
e1.pack()
 
v4=StringVar()
v4.set("Weight in Kg")
l3=Label(r1,textvariable=v4)
l3.pack()
 
v5=DoubleVar()
e2=Entry(r1,textvariable=v5,bg="black",foreground="red")
e2.pack()
 
b1=Button(r1,bg="yellow",text="Calculate",command=d2)
b1.pack()
 
v6=DoubleVar()
re=Label(r1,textvariable=v6)
re.pack()

v7=StringVar()
re1=Label(r1,textvariable=v7)

p1=PhotoImage(file="mario1.png")
p2=PhotoImage(file="mario2.png")    
p3=PhotoImage(file="mario3.png")
p4=PhotoImage(file="Breaking Bad.png")

image=Label(r1)

r1.mainloop()
```
