```python
from tkinter import *
root=Tk()
v=StringVar()
e=Entry(root,textvariable=v,insertbackground="red",
       selectbackground="green",selectforeground="black",show="*",
        insertwidth=6,insertofftime=150)
#state="disable"   or state="disabled"
e.pack()

Label(root,textvariable=v).pack()

Button(root,text="OK",state="disabled").pack()

root.mainloop()
```

```python
def d():
    e.insert(3,"333")
    
from tkinter import *
root=Tk()

e=Entry(root,insertbackground="red",
       selectbackground="green",selectforeground="black")
#state="disable"
e.pack()

pp=PhotoImage(file="sport.png")
v=StringVar()
Checkbutton(root,variable=v,text="check",offvalue="no",onvalue="yes",
           offrelief="flat",overrelief="sunken",
           selectcolor="red").pack()

l=Label(root,textvariable=v)
l.pack()

Button(root,text="OK",command=d).pack()

root.mainloop()
```
