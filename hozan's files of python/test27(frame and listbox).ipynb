```python
from tkinter import *
r=Tk()
r.geometry("400x400")

f1=Frame(r,bg="yellow")  #height=100 won't work here it will take Label height
f1.pack(side="top",fill=X)

#Label inside f1 (f1 considered as window)
#height here in Label set in lines
Label(f1,text="Student Information",font=("Time New Roman",16),bg="white",height=4).pack()

f2=Frame(r,bg="light blue")
f2.pack(side="bottom",fill=X)

Button(f2,text="Close").grid(row=0,column=0)

#width and height here in Frame in pixels
f3=Frame(r,bg="orange",width=200)     
f3.pack(side="left",fill=Y)

f4=Frame(r,bg="pink",width=200)
f4.pack(side="right",fill=Y)

r.mainloop()
```

```python
def getf():
    l.config(text=lb.get(ANCHOR))

def delf():
    lb.delete(ANCHOR)

def insf():
    tx=e.get()
    lb.insert(END,tx)
    e.delete(0,END)

from tkinter import *
r=Tk()
r.geometry("400x400")

s=StringVar()
e=Entry(r)
e.pack()

lb=Listbox(r,listvariable=s)
lb.pack()
for i in range(5):
    lb.insert(END,"text"+str(i))

Button(r,text="Get Lines",command=getf).pack(side="left")

l=Label(r)
l.pack(side="right")

Button(r,text="Delete",command=delf).pack(side="left")

Button(r,text="Insert",command=insf).pack(side="left")

r.mainloop()
```
