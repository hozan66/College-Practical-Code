```python
def dashBoard():
    root.destroy()
    global dash
    dash=Tk() #New Window
    dash.focus_force()
    dash.geometry("600x400")
    dash.title("DashBoard")
    dash.mainloop()

def submit():
    user=e1.get()
    password=e2.get()

    connection=pyodbc.connect('Driver={SQL Server};'  
                              'Server=HOZANPC;'
                              'Database=userDB;'
                              'Trusted_Connection=yes;')

    cursorObject=connection.cursor()

    query=cursorObject.execute("select * from userDB.dbo.users where userName=(?) and password=(?)",(user,password))
    if(query.fetchall()):
        messagebox.showinfo("Login","Login Successfully")
        dashBoard()
    else:
        messagebox.showerror("Login","Wrong Username or Password")
        e1.delete(0,END)
        e2.delete(0,END)

    connection.commit() 
    connection.close()
    

from tkinter import *
from tkinter import messagebox
import pyodbc
root=Tk()
root.title('Login Window')
root.focus_force()
root.resizable(False,False)
# root.geometry('550x600+10+10')
root.iconbitmap(r'student icon.ico')
root.config(bg='white')

f1=Frame(root,bg="white")
f1.pack(side='top')

f2=Frame(root,bg="white")
f2.pack(side='top',pady=16)

f3=Frame(root,bg="white")
f3.pack(side='left',pady=16)

# f4=Frame(window,bg="white")
# f4.pack(side='right')

Label(f1,text='Student Enrollment System',font=('Century',20,'bold'),bg="white").pack(pady=26)
p1=PhotoImage(file='login Enrollment.png')
Label(f1,image=p1,bg="white").pack()


Label(f1,text='Login Form',font=('Century',20,'bold'),bg="white").pack(pady=16)

Label(f1,text='User Name: ',font=('Century',18,'bold'),bg="white").pack(side='left',padx=30)
e1=Entry(f1,font=('Arial',16,'bold'),bg='#ccc')
e1.pack(side='left')

Label(f2,text='Password: ',font=('Century',18,'bold'),bg="white").pack(side='left',padx=14)
e2=Entry(f2,font=('Arial',16,'bold'),bg='#ccc',show='*')
e2.pack(side='left')

cb=PhotoImage(file="cancelB2.png")
sb=PhotoImage(file="checked2.png")
Button(f3,image=cb,bg="white",relief=FLAT,takefocus=1,command=lambda:root.destroy(),
       cursor="hand2",overrelief='ridge').pack(side='left',padx=160)
Button(f3,image=sb,bg="white",relief=FLAT,takefocus=2,command=submit,
       cursor="hand2",overrelief='ridge').pack(side='left')


root.mainloop()
```
