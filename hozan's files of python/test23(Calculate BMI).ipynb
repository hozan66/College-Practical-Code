```python
# Underweight  <  18.5
# Normalweight  18.5 - 24.9
# Overweight  24.9 - 29.9
# Obese  >=  30

# 1 Ib(pounds) = 0.4536kg
# 1 F(feet) = 0.3048M
# 0.3048M = 12 inches

# Calculate BMI (Body Mass Index)
# Where BMI equal to,
# 1- Metric
# BMI= Weight (in kg)/Height (in M)2
# 2- Standard
# BMI= (Weight (in lbs.)/ Height (in inches)     
# 2) X 703

# Note: 1 feet = 12 inches
```

```python
def f2(result):
    if(result<18.5):
        l4.config(text="Underweight")
    elif(result>=18.5 and result<=24.9):
        l4.config(text="Normalweight")
    elif(result>=25 and result<=29.9):
        l4.config(text="Overweight")
    elif(result>=30):
        l4.config(text="Obese")

def f():
    h=float(e1.get())
    w=float(e2.get())
    
    if(v.get()=="Metric"):
        result=w/(h**2)
    else:
        result=(w/(h**2)*703)

    l3.config(text=result)
    f2(result)

        
        
def g(): 
    if(v.get()=="Standard"):
        l1.config(text="Height in Inches")
        l2.config(text="Weight in Pounds")
    else:
        l1.config(text="Height in M")
        l2.config(text="Weight in KG")
        
    e1.delete(0,END)
    e2.delete(0,END)

from tkinter import *

root=Tk()
root.title("Hozan")
root.config(bg="black")

v=StringVar()
v.set("Metric")
Checkbutton(root,variable=v,textvariable=v,indicatoron=0,selectcolor="yellow",background="yellow",
            command=g,offvalue="Metric",onvalue="Standard",font=("Arial",12,"bold")).pack()


l1=Label(root,text="Height in M",bg="black",fg="white",font=("Arial",11,"bold"))
l1.pack()

e1=Entry(root,bg="#222",fg="white",bd=0,justify="center")
e1.pack()


l2=Label(root,text="Weight in KG",bg="black",fg="white",font=("Arial",11,"bold"))
l2.pack()

e2=Entry(root,bg="#222",fg="white",bd=0,justify="center")
e2.pack()

Button(root,text="Calculate",command=f,background="yellow",font=("Arial",12,"bold")).pack()

l3=Label(root,bg="black",fg="white",font=("Arial",11,"bold"))
l3.pack()
l4=Label(root,bg="black",fg="red",font=("Arial",11,"bold"))
l4.pack()

p=PhotoImage(file="cat.png")
Label(root,image=p).pack()

root.mainloop()
```

```python
from tkinter import *

window = Tk() # Create a window

#Control the label text using a stringVar
label_text = StringVar()
label_text.set("Counter : 0") # set the initial text
COUNTER = 0

def increase():
    # Function to increase the value of counter
    global COUNTER
    COUNTER += 1
    label_text.set(f"Counter : {COUNTER}")

# Create a label and set the text variable
Label(window,textvariable = label_text).pack()
Button(window,command = increase,text = "Increase").pack()

window.mainloop()
```
